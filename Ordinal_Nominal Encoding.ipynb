```python
# For ordinal encoding we need to import the sklearn preprocessing library
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
ordinalen.fit(X_train)
```

```python
X_train = ordinalen.transform(X_train)
```
